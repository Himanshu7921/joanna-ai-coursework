# Reinforcement Learning (RL) High Level Overview

## What is Reinforcement Learning?

Reinforcement Learning (RL) is a branch of Machine Learning in which an **Agent** learns how to make decisions by interacting with an **Environment**.

Unlike supervised learning, the agent is **not given the correct answers beforehand**. Instead, it learns through **trial and error**, receiving feedback in the form of **Rewards** or **Penalties**.

---

# Core Components of Reinforcement Learning

## 1. Agent

The learner or decision-maker.

Examples:
- Self-driving car
- Chess-playing AI
- Robot
- Game-playing AI

---

## 2. Environment

The world in which the agent operates.

Examples:
- Road for a self-driving car
- Chess board
- Video game world
- Grid world

---

## 3. State

The current situation of the agent within the environment.

Examples:
- Current position of a robot
- Current board configuration in chess
- Current location in a maze

---

## 4. Action

A move or decision taken by the agent.

Examples:
- Move Left
- Move Right
- Accelerate
- Brake
- Pick an object

---

## 5. Reward

A numerical feedback signal received after taking an action.

### Positive Reward (+)

Given when the action is desirable.

Examples:
- Reaching the goal
- Winning a game
- Collecting a coin

### Negative Reward / Penalty (-)

Given when the action is undesirable.

Examples:
- Hitting an obstacle
- Losing a game
- Moving away from the goal

---

# Reinforcement Learning Feedback Loop

```text
          ┌─────────┐
          │  Agent  │
          └────┬────┘
               │
            Action
               │
               ▼
       ┌──────────────┐
       │ Environment  │
       └──────┬───────┘
              │
     Reward / Penalty
              │
              ▼
       ┌──────────────┐
       │ Agent Learns │
       └──────┬───────┘
              │
              └────────────► Repeat
```

The agent continuously interacts with the environment and improves its behavior based on the feedback it receives.

---

# Learning Through Trial and Error

The fundamental idea behind RL is:

```text
Try → Receive Feedback → Learn → Improve
```

Initially, the agent knows nothing about the environment.

It explores different actions and gradually discovers which actions lead to higher rewards.

Over time, the agent develops a strategy that maximizes its total reward.

---

# Grid World Example

Consider a simple environment:

```text
A   B   C   D   E   F
┌───┬───┬───┬───┬───┬───┐
│   │   │ X │   │ X │ G │
├───┼───┼───┼───┼───┼───┤
│ X │   │   │   │   │   │
├───┼───┼───┼───┼───┼───┤
│   │ X │   │   │   │ X │
├───┼───┼───┼───┼───┼───┤
│ S │   │   │   │ X │   │
└───┴───┴───┴───┴───┴───┘
```

Where:

- **S** = Start Position
- **G** = Goal Position
- **X** = Obstacles

The agent can move:

- Up
- Down
- Left
- Right

The objective is to find a path from **Start** to **Goal** while avoiding obstacles.

---

# Exploration vs Exploitation

## Exploration

Trying new actions to gather information about the environment.

Example:

The agent tries a path it has never taken before.

---

## Exploitation

Using previously learned knowledge to maximize reward.

Example:

The agent repeatedly follows the best-known route to the goal.

---

# How Learning Happens

Suppose the agent attempts several paths:

```text
Path 1 → Hits obstacle
Path 2 → Dead end
Path 3 → Long route
Path 4 → Reaches goal
```

After many attempts:

- Good decisions receive higher rewards.
- Bad decisions receive penalties.
- The agent gradually prefers better actions.

This process allows the agent to improve without being explicitly programmed with the correct path.

---

# Objective of Reinforcement Learning

The goal is **not** to maximize immediate reward.

The goal is to maximize the **total cumulative reward** obtained over time.

Mathematically:

```text
Maximize Total Reward
=
Reward₁ + Reward₂ + Reward₃ + ...
```

A good RL agent learns actions that produce the highest long-term benefit.

---

# Self-Driving Car Analogy

Imagine a self-driving car learning to navigate a road.

### Desired Actions

- Stay within lane
- Avoid collisions
- Reach destination

Reward:

```text
+10  Safe driving
+50  Reached destination
```

---

### Undesired Actions

- Hit obstacle
- Leave lane
- Cause collision

Penalty:

```text
-10  Wrong turn
-50  Collision
```

Over thousands of driving attempts, the agent learns which actions produce higher rewards and gradually becomes better at driving.

---

# Key Intuition

Reinforcement Learning is similar to how humans learn many skills.

```text
Attempt
   ↓
Feedback
   ↓
Learning
   ↓
Improvement
```

The agent continuously interacts with its environment, learns from experience, and improves its decision-making process to maximize long-term rewards.

---

# Summary

- Reinforcement Learning is learning through interaction.
- An Agent performs actions in an Environment.
- The Environment provides Rewards or Penalties.
- The Agent learns from feedback.
- Learning occurs through trial and error.
- The objective is to maximize cumulative reward.
- Exploration discovers new possibilities.
- Exploitation uses learned knowledge.
- Over time, the agent develops an optimal strategy.

> "Reinforcement Learning is the process of learning optimal behavior through experience, feedback, and continuous interaction with the environment."
