```python
class Parent:
    def __init__(self):
        pass

    def get_hair_color(self):
        return "Black"
    
class Child(Parent):
    def __init__(self):
        pass

    def get_hair_color(self):
        return "Brown"


c = Child()
c.get_hair_color() # Black or error? # search if it belongs to child class, if yes then execut it, if no then search in Parent class, if it exits then execult the parent version
```
