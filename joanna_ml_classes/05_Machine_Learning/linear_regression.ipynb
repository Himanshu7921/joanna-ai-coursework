### Linear Regression Model

#### Step 1: Load Dataset

```python
import pandas as pd # is used for data analysis and visualization of Data
import numpy as np # is used for numeric computation

np.random.seed(42)

experience = np.arange(1, 201)  # 1 to 50 years

# Salary follows a roughly linear trend with some noise
salary = 20 + 5 * experience + np.random.randint(-10, 11, size=200)

df = pd.DataFrame({
    "Experience": experience,
    "Salary ($)": salary
})

X = df[["Experience"]]
y = df["Salary ($)"]

df.head()
```

#### Step 2: Train-Test Split

```python
train_split = int(0.8 * len(experience))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
```

#### Step 3: Load the Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

#### Step 4: Train this Model on Train set

```python
# Training the Model
model.fit(X_train, y_train)
```

#### Step 5: Test our Model on Test set

```python
x_test = np.array(X_test["Experience"])
y_test = np.array(y_test)
print(f"X_test = {x_test[:5]}")

print(f"True Values: {y_test[:5]}")

y_pred = model.predict(X_test)
print(f"Predicted Values by the Model = {y_pred}")
```

#### Step 6: Visualize Regression Line

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(
    X_test,
    y_test,
    alpha=0.5,
    label="Actual Data"
)

plt.plot(
    X_test,
    y_pred,
    color="red",
    linewidth=2,
    label="Regression Line"
)

plt.xlabel("Average Rooms")
plt.ylabel("House Price")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
