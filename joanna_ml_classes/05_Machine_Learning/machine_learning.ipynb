## 1. Introduction to Supervised vs. Unsupervised Learning

In machine learning, algorithms are broadly categorized based on how they digest dataset structures and whether they rely on ground-truth human annotations to map their objective functions.

* **Supervised Learning:** The framework provides a data matrix along with an explicit corresponding vector of human-annotated labels. The model acts as a mathematical function mapping input features directly to their designated output target.
* **Unsupervised Learning:** The framework provides raw data matrices containing only input features without any target labels. The model is forced to discover hidden internal geometric structures, feature representations, or data groupings entirely on its own.

### Supervised Learning Paradigms
Supervised pipelines are split into two primary domains depending on the topological properties of the target outputs:

1.  **Regression:** The objective is to map input data arrays to continuous numbers. The model predicts numerical real values along an infinite or bounded scale.
    * *Examples:* Salary forecasting, age prediction, student test scores, and house pricing models.
2.  **Classification:** The objective is to group data patterns into discrete qualitative classes. The network maps structural input boundaries to a categorical label space.
    * *Examples:* Email spam filtering (Class 0: Spam vs. Class 1: Not Spam) or computer vision object labeling (Class 0: Cat vs. Class 1: Dog).

## 2. Deconstructing a Regression Task: Linear Regression

Linear Regression serves as a fundamental baseline model for continuous optimization tasks. The primary goal is to find a line of best fit that passes through a cloud of data points to minimize total modeling error.

### Mathematical Formulation
To fit a single feature input $x$ to a continuous output target $y$, we establish a standard first-degree linear polynomial equation:

$$y = mx + c$$

* $y$: The dependent continuous target value we want to predict.
* $x$: The independent input feature provided by the dataset.
* $m$: The slope coefficient parameter representing the weight or sensitivity of the feature.
* $c$: The intercept constant parameter shifting the line across the vertical coordinate space (the baseline bias).

### The Step-by-Step Optimization Workflow

* **Step 1: Define the Baseline Hypothesis Space** We establish our model's linear architecture ($y = mx + c$) and initialize parameters $m$ and $c$ with arbitrary random values.
* **Step 2: Calculate Predictive Output Fit** We pass our dataset features through the current linear parameters to compute predicted points ($y_p = mx + c$) for every single observation in the collection.
* **Step 3: Evaluate Predictive Accuracy via Loss Functions** We measure the error between our model’s predictions ($y_p$) and actual human-validated targets ($Y_t$) using an objective metric called the Mean Squared Error (MSE) loss:
    
    $$L = \frac{1}{N} \sum_{i=1}^{N} (Y_t - Y_p)_i^2$$
    
* **Step 4: Minimize Loss via Parameter Optimization** Using optimization strategies like Gradient Descent, we calculate partial derivatives ($\frac{\partial L}{\partial m}$, $\frac{\partial L}{\partial c}$) to determine how the loss changes relative to our weights. We then update our weights in the opposite direction of the gradient using a designated learning rate ($\alpha$):
    
    $$m^* \leftarrow m - \frac{\partial L}{\partial m} \cdot \alpha$$
    $$c^* \leftarrow c - \frac{\partial L}{\partial c} \cdot \alpha$$
    
* **Step 5: Execute Final Model Predictions** Once parameters converge to their optimized targets ($m^*, c^*$), the loss reaches its global minimum. The final stable equation can now calculate inference predictions for entirely new data points.

### Real-World Execution Trace
Consider a tracking system measuring student scores against study hours per week:

| Hours Studied ($x$) | Actual Score ($y$) | Linear Fit Execution ($y = 5.3x + 37$) | Final Predicted Score Output |
| :--- | :--- | :--- | :--- |
| 1 | 40 | $y = 5.3(1) + 37 = 42.3$ | $\approx 42$ |
| 2 | 48 | $y = 5.3(2) + 37 = 47.6$ | $\approx 48$ |
| 3 | 52 | $y = 5.3(3) + 37 = 52.9$ | $\approx 53$ |
| 4 | 60 | $y = 5.3(4) + 37 = 58.2$ | $\approx 58$ |
| 10 | 90 | $y = 5.3(10) + 37 = 90.0$ | $\approx 90$ |
| **11 (Inference)** | **Unknown** | **$y = 5.3(11) + 37 = 95.3$** | **$\approx 95$** |

Using optimization software like Desmos, the parameters stabilize at $m^* = 5.3$ and $c^* = 37.3$. When a student studies for 11 hours, the model uses these calibrated parameters to evaluate $5.3(11) + 37.3$ and successfully predicts a final score of approximately 95.

## 3. Classifying Supervised Algorithms

Supervised architectures are chosen based on data dimensions, spatial distribution complexity, and structural linearity constraints.

### Linear & Statistical Supervised Baselines
* **Linear Regression:** Fits continuous geometric target spaces through multi-dimensional feature point clouds.
* **Support Vector Machines (SVM):** Constructs optimized multi-dimensional hyperplanes to maximize geometric separation boundaries between different classes.
* **Decision Trees:** Builds a hierarchical flowchart-style tree structure based on feature splitting thresholds to make step-by-step predictions.
* **Naïve Bayes Theorem:** A classification technique based on Bayes' Theorem that calculates conditional probability spaces under strict independent feature assumptions.
* **K-Nearest Neighbors (KNN):** A non-parametric classifier that assigns a label to an unlabeled data point by evaluating the majority vote of its closest neighboring points in a spatial coordinate plane.
* **Logistic Regression:** Maps input features to a discrete binary classification landscape using a single activation equation wrapped inside a Sigmoid function:
    
    $$y = \sigma(Wx + b)$$

### Transitioning to Deep Neural Networks (MLP)
When datasets present complex, non-linear relationships that simple regression pipelines cannot resolve, we stack multiple linear transformations together. 

A **Multi-Layer Perceptron (MLP)** scales up the concept of Logistic Regression by stacking multiple transformation layers together and passing the outputs through non-linear activation functions (such as Softmax for multi-class classification):

$$y = \text{Softmax}(W_n( \dots \sigma(W_1x + b_1) \dots ) + b_n)$$

By connecting these mathematical layers in a dense network, the model can capture highly complex features, laying the foundation for modern Deep Learning.
