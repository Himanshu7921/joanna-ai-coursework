## Lecture-01: Gradient Descent

````python
import matplotlib.pyplot as plt
import numpy as np

# --------------------------------------------------
# A very simple loss function
# Think of this as the error made by a model
# Our goal will be to find the minimum loss
# --------------------------------------------------
def f(x):
    return x ** 2


print("Creating values of theta from -1 to 1...")

# Generate different values of theta (model parameter)
xs = np.arange(-1, 1, 0.01)

print(f"Generated {len(xs)} theta values.")

print("Calculating loss for each theta value using f(x) = x²...")

# Calculate corresponding loss values
y = f(xs)
print("Plotting Loss vs Theta...")

# Create the graph
plt.plot(xs, y)
plt.xlabel("Theta (Parameter)")
plt.ylabel("Loss")
plt.title("Loss Landscape")
plt.show()

print("Notice that the minimum loss occurs at theta = 0.")
print("This is exactly what Gradient Descent tries to find automatically.")
````

````python
# A very tiny number
# Used to make a very small change in theta
h = 0.00000001

# Current parameter value of the model
# Think of theta as a weight that the model is trying to learn
theta = 0.75

# Calculate the loss at theta = 0.75
f(theta)
````

## Gradient Descent Algorithm

````python
print("Starting Gradient Descent...\n")
for step in range(1000):
    # Compute gradient at current theta
    dtheta = (f(theta + h) - f(theta)) / h
    # Update theta
    theta = theta - 0.0001 * dtheta
    if step % 100 == 0:
        print(
            f"Step {step:4d} | "
            f"Gradient = {dtheta:.6f} | "
            f"Theta = {theta:.6f} | "
            f"Loss = {f(theta):.6f}"
        )
print("\nGradient Descent Finished!")
print(f"Best theta found = {theta}")
print(f"Final loss       = {f(theta)}")


# 1. The model starts with a random parameter value (theta).
# 2. We calculate the gradient.
#    The gradient tells us which direction increases the loss.
# 3. We move in the opposite direction:
#    theta = theta - learning_rate * gradient
# 4. After the update, the loss becomes smaller.
# 5. We repeat this process many times.
# 6. Eventually theta reaches the minimum point of the loss curve.

# This simple loop is the heart of modern AI.
    # Whether it is:
    # - CNNs
    # - Transformers
    # - GPT
    # - U-Net
    # - Diffusion Models

# all of them repeatedly do:

# Forward Pass
# → Compute Loss
# → Compute Gradient
# → Update Parameters
# → Repeat
````

## Next Lecture: Moving from Scalars to Tensors

In this lecture, we learned how a model can learn using a single parameter:

$$
y = w x + b
$$

where both the input and the parameters were simple scalars.

This is useful for understanding **Gradient Descent**, but real-world AI systems never work with a single number.

A cat image is not a scalar.

A sentence is not a scalar.

A video is not a scalar.

Modern AI systems work with **vectors, matrices, and tensors**.

---

### Why Move Beyond Scalars?

Imagine trying to classify a cat using only one number:

$$
x = 5
$$

There is almost no information in a single scalar.

The model can only learn a very simple relationship.

With tensors, we can represent much richer information:

$$
X \in \mathbb{R}^{3 \times H \times W}
$$

where an image contains thousands of pixel values instead of a single number.

Now the model can learn:

* Edges
* Shapes
* Colors
* Textures
* Object Parts
* High-Level Concepts

instead of learning from a single scalar value.

---

### More Expressive Weights and Biases

Previously:

$$
y = wx + b
$$

Here:

* One weight
* One bias
* One input

The model can only learn a very limited mapping.

With tensors:

$$
Y = WX + b
$$

where:

* $X$ is a tensor
* $W$ is a tensor
* $b$ is a tensor

Now the model can learn thousands or millions of relationships simultaneously.

Instead of asking:

> "How does one number affect another number?"

the model learns:

> "How do thousands of features interact with each other?"

This dramatically increases the learning capacity of the network.

---

### Why Tensors are Faster

Tensors are not only more expressive but also computationally efficient.

Modern hardware such as GPUs is designed to perform operations on entire tensors at once.

Instead of:

```text
Process one value
Process one value
Process one value
...
```

we perform:

```text
Process millions of values in parallel
```

This is called **vectorized computation**.

As a result:

* Faster training
* Better GPU utilization
* Lower execution overhead
* Massive scalability

---

### The Bigger Picture

A single neuron learns:

$$
y = wx + b
$$

A neural network learns:

$$
Y = f(X;\theta)
$$

where:

* $X$ becomes a tensor
* $\theta$ becomes millions of weights and biases
* Gradient Descent optimizes all of them together

This transition from **scalars → tensors** is what allows modern AI systems such as CNNs, Transformers, U-Nets, GPTs, and Diffusion Models to learn complex patterns from massive amounts of data.

> Scalars help us understand learning. Tensors make large-scale learning possible.
