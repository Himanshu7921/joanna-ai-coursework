### Mathematical Reference for Gradients
Given the MSE Loss function:
$$L = \frac{1}{N} \sum_{i=1}^{N} (y_i - (m \cdot x_i + c))^2$$

The partial derivatives (gradients) with respect to parameters $m$ and $c$ are:
$$\frac{\partial L}{\partial m} = -\frac{2}{N} \sum_{i=1}^{N} x_i \cdot (y_i - y_{pred\_i})$$
$$\frac{\partial L}{\partial c} = -\frac{2}{N} \sum_{i=1}^{N} (y_i - y_{pred\_i})$$
