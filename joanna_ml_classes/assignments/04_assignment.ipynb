## Overview

In this assignment, you will implement a simplified Transformer inference pipeline capable of predicting the next token from a given sequence. Your implementation should closely resemble the data flow of modern autoregressive language models while remaining lightweight enough to be implemented using only NumPy.

Your solution should emphasize modularity, readability, and correctness.

## Requirements

### Self-Attention Module

Implement a reusable function that computes the self-attention output given the Query (`Q`), Key (`K`), and Value (`V`) matrices.

Your implementation should:

* Compute the attention score matrix.
* Normalize the scores using the Softmax function.
* Produce the attention output.
* Support arbitrary sequence lengths and embedding dimensions without requiring modifications to the implementation.

---

### Prediction Module

Implement a prediction function that receives the final hidden representation, projects it into the vocabulary space using a learnable weight matrix and bias vector, and predicts the next token.

Use only the final hidden representation for inference.

The vocabulary should contain at least eight unique tokens of your choice.

---

### Generalization

Your implementation must not assume a fixed number of tokens.

Demonstrate that the same implementation works correctly for multiple input sequence lengths without modifying the source code.

Examples include, but are not limited to:

* 3 tokens

### Code Organization

Structure your implementation into reusable functions.

Avoid writing the entire pipeline inside a single code cell or function.

Each function should have a single well-defined responsibility.
