#### Why Hallucinations occur?
An LLM doesn't know what's true it only predicts a probability distribution over the next token. Softmax assigns a probability to every token in the vocabulary, and the decoding algorithm (greedy or sampling) chooses the next token from this distribution. Hallucinations occur because the model is optimized to predict the most likely continuation of text, not to verify factual correctness.

## Let's code mini Toy GPT for example

```python
import numpy as np
# ------------------ Defining the Query, Key and Value Vectors for Attention Mechanism -----------------
Q = np.random.rand(4, 2)
K = np.random.rand(4, 2)
V = np.random.rand(4, 2)

# ------------------ Defining the Weights and biases for MLP -----------------
W = np.random.rand(2, 5)
b = np.random.rand(5, )
```

```python
# ----------- Utility funcitons ------------------------
def softmax(Z):
    return np.exp(Z) / np.sum(np.exp(Z), axis = -1, keepdims=True) 

def compute_attention_scores(Q, K, V):
    Z = softmax(Q @ K.T) # Q.shape = (4, 2) @ K.shape = (4, 2).T = (2, 4) --> (4, 2) @ (2, 4) = (4, 4)
    attention_score = Z @ V # (4, 2)
    return attention_score

def MLP(x, W, b):
    return x @ W + b
```

```python
attention_score = compute_attention_scores(Q, K, V)
available_tokens = [".", ",", ";", "the", "is"] # 5 tokens for out toy example
# -------------------------------------------------------------------------------------------------------------------------------------------
y = MLP(attention_score, W, b)
# probs = softmax(y.sum(axis=0)) # Instead of just mixing all the infromation we can just use the last representation
probs = softmax(y[-1]) # Instead of just mixing all the infromation we can just use the last representation [this is what real GPT does]

model_input = "The Cat drank milk"
print(f"{model_input} ------> [GPT-Model] --------> [Output: '{available_tokens[probs.argmax()]}']")
```
